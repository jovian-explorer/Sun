```python
#-----------------------------------------------------------------------------------------------------------------
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
#-----------------------------------------------------------------------------------------------------------------
plt.figure(figsize=(8,4),dpi=100)

a =356
b = 0
c= 0.105

rad =  (0.03218077540106952,0.02683281417112299,0.0376682486631016,0.024368101604278075,0.029437038770053475,0.03483150401069519,1.0)
r = np.arange(0,1.25,0.01)
r = np.array(r)
v = (112.8371379956578,126.64897552013602,172.2372178386981,189.5309413834875,212.84972355776773,260.60766492906432,351.13)

velo = np.sqrt( (a**2) * (1- np.exp(-(r-b)/c)))

# print(velo)
plt.scatter(rad,v,s=15,label = "Data")
plt.plot(r,velo,'r--',label = "Curvefit")
plt.ylim(75,425)

plt.xlabel("$Radius (R_{\odot}) \longrightarrow$")
plt.ylabel("$V_{km/s} \longrightarrow $")
plt.legend(loc = 'upper left')
plt.show()
```
