```python
#-----------------------------------------------------------------------------------------------------------------
#importing libraries
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
#-----------------------------------------------------------------------------------------------------------------
#importing data from my GitHub
url = 'https://raw.githubusercontent.com/jovian-explorer/Sun/main/MOM/OMNI/ssn_number.txt'
pd.set_option('display.max_columns', None) 
pd.set_option('display.max_rows', None)  
pd.set_option('display.max_colwidth', None)
sunspot_number = pd.read_csv(url,delimiter = '\t')
print(sunspot_number.head(5))
sunspot_number.columns = ["YEAR","DOY","SSN"]
YEAR = np.array(sunspot_number.YEAR)	        # Values of second_moment
DOY = np.array(sunspot_number.DOY)
SSN = np.array(sunspot_number.SSN)	
# -----------------------------------------------------------------------------------------------------------------

fig, ax = plt.subplots(figsize=(8,4),dpi=800)
plt.plot(YEAR+(DOY/365),SSN,'k-',linewidth=1)
# plt.scatter(YEAR+(DOY/365),SSN,color='k',s=1)

plt.axvline(x=1991,color = 'k')
plt.axvline(x=2000,color = 'k')
plt.axvline(x=2015,color = 'k')
plt.axvline(x=2021,color = 'k')

plt.xlim(1985,2022.85)

plt.xlabel("Year")
plt.ylabel("SSN Number")

plt.show()
# -----------------------------------------------------------------------------------------------------------------
fig.savefig("/home/dev/Sun/Stage_1_Project/Images/ssn.jpg")
```
